# Introduction to Pandas Module

```python
import pandas as pd

# people = P1, P2, P3,
# Age = 30, 35, 40,

dict = {"People": ["P1", "P2", "P3"], "Age": [30, 35, 40], "Salary": [200000, 250000, 300000]}

df = pd.DataFrame(dict)
print(df)
```

```python
# df["Key"].operation()
d1 = df["Salary"].sum()
print(f"Sum of Salary: {d1}")

d2 = df["Age"].sum()
print(f"Sum of Age: {d2}")

d3 = df["Salary"].mean()
print(f"Mean of Salary: {d3}")
```

### More Pandas Operations

```python
# Filtering data: Find people with Age greater than 30
filtered_df = df[df['Age'] > 30]
print("People with Age greater than 30:")
print(filtered_df)
```

```python
# Sorting data: Sort by Salary in descending order
sorted_df = df.sort_values(by='Salary', ascending=False)
print("Data sorted by Salary (descending):")
print(sorted_df)
```

```python
# Adding a new column: Let's assume everyone gets a bonus of 10% of their salary
df['Bonus'] = df['Salary'] * 0.10
print("DataFrame with new 'Bonus' column:")
print(df)
```

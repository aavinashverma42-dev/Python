# Introduction to Numpy


```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])
print(arr)
```

Basic ooperation on Numpy

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

arr3 = np.concatenate((arr1, arr2))
print(arr3)

arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[5, 6], [7, 8]])

arr3 = np.concatenate((arr1, arr2), axis=1)
print(arr3)
```

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

arr3 = np.stack((arr1, arr2), axis=1)
print(arr3)
```

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

print(arr1 + arr2)
print(arr1 - arr2)
print(arr1 * arr2)
print(arr1 / arr2)
arr1 +arr2
```

```python
# Indexing
arr = np.array([1, 2, 3, 4, 5])
print(arr[0])
print(arr[1])
print(arr[2])
print(arr[3])
print(arr[4])
```

```python
# Slicing
arr = np.array([1, 2, 3, 4, 5])
print(arr[0:3])
print(arr[1:4])
print(arr[2:5])
print(arr[3:6])
```

```python
# matrix shaping
arr1 = np.array([1, 2, 3, 4, 5, 6])

np.shape(arr1)
# output format = (no. of rows, no. of columns)
```

```python
arr1 = np.array([1, 2, 3, 4, 5, 6])
arr1.reshape((2,3))
```

```python
m1 = np.array([[1, 2], [4, 5]])
m2 = np.array([[7, 8], [10, 11]])

print(m1 + m2)
m3 = np.dot(m1, m2)
print(m3)
```

```python
arr1 = np.array([1, 2, 3, 4, 5, 6])
np.mean(arr1)
```

```python
arr1 = np.array([1, 2, 3, 4, 5, 6])
np.std(arr1)
```

# Working with String

```python
# string = sequence of characters, immutable..
double_quote_string = "Avinash"
single_quote_string = 'Avinash'
tripple_quote_string = '''Avinash'''
```

```python
multi_line_string = '''I am a good python
programmer doing programming all time.
'''
print(multi_line_string)
```

# Operations on string

```python
# // Concatenation
a = "Hello"
b = "World"
print(a + " " + b)
```

```python
# // Repetition
c = "abcd"
print(c * 3)
```

```python
# // Indexing
my_string = "Python"

# forward indexing
print(my_string[0])
print(my_string[1])
print(my_string[2])
print(my_string[3])
print(my_string[4])
print(my_string[5])

# reverse indexing
print(my_string[-1])
print(my_string[-2])
print(my_string[-3])
print(my_string[-4])
print(my_string[-5])
print(my_string[-6])

# Slicing
print(my_string[0:6])
print(my_string[0:5])
print(my_string[0:4])
print(my_string[0::2])
```

```python
# // length
my_string = "Python"
print(len(my_string))
```

```python
# // Membership
my_string = "Python"
print('P' in my_string)
print('T' in my_string)
print('P' not in my_string)
print('n' not in my_string)
print('p' in my_string.lower())
```

```python
# //Case conversion
my_string = "Python world"
print(my_string.upper())
print(my_string.lower())
print(my_string.title())
print(my_string.swapcase())
print(my_string.capitalize())    # alpha>>Alpha
```

```python
# // Stripping
my_string = "   ---Python---   "
print(my_string.strip())
print(my_string.lstrip())
print(my_string.rstrip())
```

```python
# // Splitting
my_string = "Python-is-a-programming-language"
print(my_string.split('-'))
print(my_string.split('a'))
print(my_string.split('is'))
```

```python
# // Joining
my_string = "apple","banana","cherry"
print(",".join(my_string))
```

```python
# // Replacing
a = "Hello World"
print(a.replace("World","Python"))
```

```python
# // Finding
a = "Hello World"
print(a.find("o"))   # give the index of the desired characters
print(a.index('l'))

b = "Banana"
print(b.count("a"))
```

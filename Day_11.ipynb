# File Handling

Creating a txt file....

```python
file_name = 'my_file.txt'
content = 'Hello, this is a test file created in Colab.'

with open(file_name, 'w') as f:
    f.write(content)

print(f"File '{file_name}' created successfully with content:\n'{content}'")
```

Opening the file in read mode and printing the content within it...

```python
with open("my_file.txt","r") as f:
    content = f.read()
    print(content)
```

opening the file in append mode and printing the content within it after the append mode execution....

```python
with open("my_file.txt","a") as f:
    content = f.write("\nThis the third line....")
with open("my_file.txt","r") as f:
    content = f.read()
    print(content)
```

Creating a csv file.....

```python
import csv

file_name = 'my_csv.csv'

with open(file_name, 'w', newline='') as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(["col1", "col2"])
    csv_writer.writerow(["data1", "data2"])
```

## What are Python Modules?

In Python, a module is essentially a file containing Python definitions and statements. The filename is the module name with the suffix `.py`.

Modules allow you to logically organize your Python code. They provide a way to break down large programs into smaller, manageable, and reusable files. This makes code more readable, maintainable, and efficient.


```python
# Import the math module and use it to find the square root of 25

import math
# import math as m

a = math.sqrt(25)     # math----->>> m...
print(a)            # a = m.sqrt(25)

a1 = math.pi
print(a1)

b = math.pow(2, 10)
print(b)

b1 = math.floor(2.9)
print(b1)

c = math.ceil(2.1)
print(c)

c1 = math.factorial(5)
print(c1)
```

```python
import datetime

# Get the current date and time
current_datetime = datetime.datetime.now()
print(f"Current date and time: {current_datetime}")

# Get the current date
current_date = datetime.date.today()
print(f"Current date: {current_date}")

# Get the current time (without date)
current_time = datetime.datetime.now().time()
print(f"Current time: {current_time}")

# Create a specific date
specific_date = datetime.date(2023, 10, 26)
print(f"Specific date: {specific_date}")

# Create a specific datetime
specific_datetime = datetime.datetime(2023, 10, 26, 14, 30, 0)
print(f"Specific datetime: {specific_datetime}")
```

```python
# Accessing components of a datetime object
print(f"Year: {current_datetime.year}")
print(f"Month: {current_datetime.month}")
print(f"Day: {current_datetime.day}")
print(f"Hour: {current_datetime.hour}")
print(f"Minute: {current_datetime.minute}")
print(f"Second: {current_datetime.second}")
print(f"Weekday (0=Monday, 6=Sunday): {current_datetime.weekday()}")
```

```python
# Formatting dates and times using strftime()
# Common format codes:
# %Y: Year with century (e.g., 2023)
# %m: Month as a zero-padded decimal number (01-12)
# %d: Day of the month as a zero-padded decimal number (01-31)
# %H: Hour (24-hour clock) as a zero-padded decimal number (00-23)
# %M: Minute as a zero-padded decimal number (00-59)
# %S: Second as a zero-padded decimal number (00-59)
# %A: Weekday as full name (e.g., 'Thursday')
# %B: Month as full name (e.g., 'October')

formatted_date_time = current_datetime.strftime("%Y-%m-%d %H:%M:%S")
print(f"Formatted date and time: {formatted_date_time}")

custom_format = current_datetime.strftime("Today is %A, %B %d, %Y at %I:%M %p")
print(f"Custom formatted string: {custom_format}")
```

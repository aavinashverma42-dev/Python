# `Functions with default parameters`

```python
# Normally
def hello(name):
  print(f"Hello, {name}")
hello("Avinash")

# Default
def hello(name="Avinash"):
  print(f"Hello,{name}")
hello()
hello("Sir")   # you can pass another name...
```

# Keyword Arguments

```python
# add two numbers

def sum(a=10,b=20):
  return a + b

sum()    #output = 30
sum(40,60)     # Output = 100
```

# Global and Local Variables

```python
#Global variables = defined outside of a function
x = 101

def test(y):
  global x
  print(x, y)

test(10)

# Local variables = defined inside of a function

def test(x):
   x = 20
   print(x)

test(x)
```

# Nested Functions

```python
# Functions inside a function

def hello(name):
  print(f"Hello, {name}")

def fav_city(city_name,name):
  hello(name)
  print(f"Welcome to {city_name}!")

fav_city("Mumbai","Avinash")
```

# INTRODUCTION TO SETS

### Common Set Operations

Sets are powerful for tasks involving unique collections of items. Here are the most common operations:

1.  **Adding Elements**: Use the `add()` method to add a single element, or `update()` to add multiple elements from an iterable.
2.  **Removing Elements**: Use `remove()`, `discard()`, `pop()`, or `clear()`.
    *   `remove(element)`: Removes the specified element. Raises a `KeyError` if the element is not found.
    *   `discard(element)`: Removes the specified element if it is present. Does nothing if the element is not found.
    *   `pop()`: Removes and returns an arbitrary element from the set.
    *   `clear()`: Removes all elements from the set.
3.  **Set Mathematical Operations**:
    *   **Union**: Combines elements from two or more sets ( `|` operator or `union()` method).
    *   **Intersection**: Returns elements common to all sets (`&` operator or `intersection()` method).
    *   **Difference**: Returns elements in the first set but not in the second (`-` operator or `difference()` method).
    *   **Symmetric Difference**: Returns elements that are in either set, but not in both (`^` operator or `symmetric_difference()` method).
4.  **Membership**: Check if an element exists in a set using the `in` operator.
5.  **Length**: Determine the number of elements in a set using the `len()` function.
6.  **Subsets and Supersets**: Check relationships between sets (`issubset()`, `issuperset()`).

```python
# Initialize a set
my_set = {1, 2, 3}
print(f"Initial Set: {my_set}")
```

Operation on Set......

```python
# 1. Adding Elements
print("\n--- Adding Elements ---")
my_set.add(4) # Add a single element
print(f"After adding 4: {my_set}")

my_set.add(2) # Adding an existing element has no effect
print(f"After adding 2 (again): {my_set}")

my_set.update([5, 6, 7]) # Add multiple elements from an iterable
print(f"After updating with [5, 6, 7]: {my_set}")
```

```python
# 2. Removing Elements
print("\n--- Removing Elements ---")
my_set.remove(7) # Remove an existing element
print(f"After removing 7: {my_set}")

my_set.discard(10) # Discard a non-existing element (no error)
print(f"After discarding 10: {my_set}")

my_set.discard(6) # Discard an existing element
print(f"After discarding 6: {my_set}")

popped_element = my_set.pop() # Remove and return an arbitrary element
print(f"After popping: {my_set}, Popped element: {popped_element}")

# Let's re-add some elements for later set operations
my_set.update([1, 8, 9])
print(f"Set for further operations: {my_set}")
```

```python
# 3. Set Mathematical Operations
print("\n--- Set Mathematical Operations ---")
set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}
print(f"Set A: {set_a}")
print(f"Set B: {set_b}")

# Union
print(f"Union (A | B): {set_a | set_b}")
print(f"Union (A.union(B)): {set_a.union(set_b)}")

# Intersection
print(f"Intersection (A & B): {set_a & set_b}")
print(f"Intersection (A.intersection(B)): {set_a.intersection(set_b)}")

# Difference
print(f"Difference (A - B): {set_a - set_b}")
print(f"Difference (A.difference(B)): {set_a.difference(set_b)}")

# Symmetric Difference
print(f"Symmetric Difference (A ^ B): {set_a ^ set_b}")
print(f"Symmetric Difference (A.symmetric_difference(B)): {set_a.symmetric_difference(set_b)}")
```

```python
# 4. Membership
print("\n--- Membership ---")
print(f"Is 2 in set_a? {2 in set_a}")
print(f"Is 7 in set_a? {7 in set_a}")
```

```python
# 5. Length
print("\n--- Length ---")
print(f"Length of my_set: {len(my_set)}")
```

```python
# 6. Subsets and Supersets
print("\n--- Subsets and Supersets ---")
set_c = {1, 2}
print(f"Set C: {set_c}")
print(f"{set_c.issubset(set_a)}")
print(f"{set_a.issuperset(set_c)}")
```

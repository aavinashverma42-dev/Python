```python
# For loop

for i in range(1, 11):
  print(f"The number is {i}")

print("\n")

num = int(input("Enter a number: "))

if num > 1:
    for i in range(2, num):
        if (num % i) == 0:
            print(num, "is not a prime number")
            break
    else:
        print(num, "is a prime number")
```

```python
# While loop

num = int(input("Enter a number: "))
i = 1
while i <= num:
  print(f"The number is {i}")
  i += 1
```

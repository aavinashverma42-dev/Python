Ques.. Lets say we have a list l1 = [1,2,2,3,2,4,3,5]..
       Identify the most repeated number in the list.

```python

l1= [1, 2, 2, 3, 2, 4, 3, 5]
frequency = {}

for item in l1:
    if item in frequency:
       frequency[item] += 1
    else:
       frequency[item] = 1

most_repeated = None
max_count = 0

for key, value in frequency.items():
  if value > max_count:
    max_count = value
    most_repeated = key

print("Frequency", frequency)
print("Most repeated number", most_repeated)
print("Occurrence", max_count)
```

### Common Tuple Operations

Tuples, being immutable, have a more limited set of operations compared to lists or dictionaries. Here are the most common ones:

1.  **Accessing Elements**: Retrieve individual elements or a range of elements using indexing and slicing.
2.  **Concatenation**: Combine two or more tuples using the `+` operator.
3.  **Repetition**: Repeat a tuple multiple times using the `*` operator.
4.  **Length**: Determine the number of elements in a tuple using the `len()` function.
5.  **Membership**: Check if an element exists in a tuple using the `in` operator.
6.  **Iteration**: Loop through the elements of a tuple.
7.  **Finding Minimum/Maximum**: Get the smallest or largest element (if comparable) using `min()` and `max()`.
8.  **Counting Elements**: Count the occurrences of a specific element using the `count()` method.
9.  **Finding Index**: Find the index of the first occurrence of a specific element using the `index()` method.

Keep in mind that operations like adding, modifying, or deleting individual elements are not possible with tuples directly.

```python
my_tuple = (10, 20, 30, 40, 20)
print(f"Original Tuple: {my_tuple}")

# 1. Accessing Elements (Indexing and Slicing)
print("\n--- Accessing Elements ---")
print(f"First element (index 0): {my_tuple[0]}")
print(f"Last element (negative indexing): {my_tuple[-1]}")
print(f"Elements from index 1 to 3 (exclusive): {my_tuple[1:4]}")
```

```python
# 2. Concatenation
print("\n--- Concatenation ---")
tuple_b = (50, 60)
combined_tuple = my_tuple + tuple_b
print(f"Combined Tuple: {combined_tuple}")
```

```python
# 3. Repetition
print("\n--- Repetition ---")
repeated_tuple = my_tuple * 2
print(f"Repeated Tuple (x2): {repeated_tuple}")
```

```python
# 4. Length
print("\n--- Length ---")
print(f"Length of my_tuple: {len(my_tuple)}")
```

```python
# 5. Membership
print("\n--- Membership ---")
print(f"Is 30 in my_tuple? {30 in my_tuple}")
print(f"Is 99 in my_tuple? {99 in my_tuple}")
```

```python
# 6. Iteration
print("\n--- Iteration ---")
print("Iterating through my_tuple:")
for item in my_tuple:
    print(item)
```

```python
# 7. Finding Minimum/Maximum (for comparable elements)
print("\n--- Min/Max ---")
print(f"Minimum element: {min(my_tuple)}")
print(f"Maximum element: {max(my_tuple)}")
```

```python
# 8. Counting Elements
print("\n--- Counting Elements ---")
print(f"Count of 20 in my_tuple: {my_tuple.count(20)}")
print(f"Count of 10 in my_tuple: {my_tuple.count(10)}")
```

```python
# 9. Finding Index
print("\n--- Finding Index ---")
print(f"Index of first occurrence of 30: {my_tuple.index(30)}")
```

# Introduction to Dictionaries

Creating a Dictionary


A **dictionary** in Python is an unordered collection of data values, holds **key:value** pairs. Keys must be unique and immutable (e.g., strings, numbers, or tuples), while values can be any data type and can be duplicated.

```python
student = {"name": "Raju", "age": 23, "gender": "male"}
print(student)
```

Accessing Values

```python
student["age"]
student["name"]
```

```python
# Update
student["age"] = 30
print(student)
```

LOOPING THROUGH DICTIONARIES

```python
student = {"name": "Raju", "age": 23, "gender": "male"}

for key in student:
  print(key)
print("---------")

for key in student:
  print(key,"--", student[key])
```

### Common Dictionary Operations

Here are the most common operations you can perform on Python dictionaries:

1.  **Accessing Values**: Retrieve a value using its key.
2.  **Adding/Updating Elements**: Add a new key-value pair or update an existing one.
3.  **Removing Elements**: Delete key-value pairs.
4.  **Getting Keys, Values, and Items**: Extract different components of the dictionary.
5.  **Checking for Key Existence**: Verify if a key is present in the dictionary.
6.  **Getting Dictionary Size**: Determine the number of key-value pairs.

```python
# Initialize a dictionary for demonstration
my_dict = {'fruit': 'apple', 'color': 'red', 'count': 3}
print(f"Initial Dictionary: {my_dict}")

# Removing Elements
print("\n--- Removing Elements ---")
del my_dict['color']
print(f"{my_dict}")

# Using 'pop()' to remove an item and get its value
popped_count = my_dict.pop('count')
print(f"{my_dict}, Popped value: {popped_count}")

# Add an item back to demonstrate popitem
my_dict['new_item'] = 'value'
print(f"{my_dict}")

popped_item = my_dict.popitem() # Removes the last inserted item
print(f"\n{my_dict}, Popped item: {popped_item}")

# Using 'clear()' to remove all items
my_dict.clear()
print(f"\nAfter 'my_dict.clear()': {my_dict}")
```

```python
# 4. Getting Keys, Values, and Items
print("\n--- Getting Keys, Values, and Items ---")

print(f"Keys: {list(my_dict.keys())}")
print(f"Values: {list(my_dict.values())}")
print(f"Items: {list(my_dict.items())}")
```

```python
# 5. Checking for Key Existence
print("\n--- Checking for Key Existence ---")

print(f"{'fruit' in my_dict}")
print(f"{'taste' in my_dict}")
```

```python
# 6. Getting Dictionary Size
print("\n--- Getting Dictionary Size ---")

print(f"Size of dictionary: {len(my_dict)}")
```

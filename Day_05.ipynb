```python
# Working with Lists

# A list is a built-in data type in Python used to store multiple items in a single variable.
# Lists are created using square brackets: []

fruits = ["apple", "banana", "cherry", "apple"]
display(fruits)

```

### Common List Operations

Lists support various methods to manipulate their contents. Here are the most frequent ones:

```python
# 1. Adding Elements
fruits.append("orange")          # Adds to the end
fruits.insert(1, "mango")        # Adds at a specific index

more_fruits = ["grape", "kiwi"]
fruits.extend(more_fruits)       # Appends elements from another list

print("After adding elements:")
display(fruits)
```

```python
# 2. Removing Elements
fruits.remove("apple")           # Removes the first occurrence of 'apple'
popped_item = fruits.pop()      # Removes and returns the last item

print(f"Removed item: {popped_item}")
print("After removing elements:")
display(fruits)
```

```python
# 3. Organizing a List
fruits.sort()                   # Sorts the list alphabetically (in-place)
print("Sorted list:")
display(fruits)

fruits.reverse()                # Reverses the order of the list
print("Reversed list:")
display(fruits)
```

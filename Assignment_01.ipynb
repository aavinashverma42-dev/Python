```python
#Daily task organizer

#Create a empty checklist
checklist = []

# Lists to store completed and incompleted tasks
completed_tasks = []
incomplete_tasks = []

# input the numbers of Tasks
n = int(input("Enter the number of Tasks for Today: "))

#ADD tasks to the checklist
for i in range(n):
  task = input(f"Enter task:{i + 1} ")
  checklist.append(task)

print("-\n--- End of the DAY Review ---")

# Check wether each task is completed
for task in checklist:
  status = input(f"Is '{task}' completed? (yes/no): ")
  if status.lower() == "yes":
    completed_tasks.append(task)
  else:
    incomplete_tasks.append(task)

  # Display the results
print("\n--- End of the Daily Routine ---")

print("\n--- Completed Tasks ---")
if completed_tasks:
  for tasks in completed_tasks:
    print("-", tasks)
else:
  print("No completed tasks.")

print("\n--- Incomplete Tasks ---")
if incomplete_tasks:
  for tasks in incomplete_tasks:
    print("-", tasks)
else:
  print("No incomplete tasks.")
```

# Handling all those Error

Using Try Except Block

```python
try:
  a = 10
  b = a / 0
  print(b)
except ZeroDivisionError as e:
  print("Error: Cannot divide by zero")
```

```python
try:
  a = int("hello")
  print(a)
except ValueError as e:
  print("Error: Invalid Value")
```

Using Try Except and Finally Block

```python
try:
  a = int("hello")
  print(a)
except ValueError as e:
  print("Error: Invalid Value")
finally:
  print(2+3)
```
